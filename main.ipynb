```python
import pandas as pd
import math
```

```python
df = pd.read_csv('acoes.csv')
df = df.set_index('Date')
df.index = pd.to_datetime(df.index, format='%Y-%m-%d')
df
```

```python
daily_return = df.pct_change()
daily_return
```

```python
cummulative_return = (1 + daily_return).cumprod() - 1
cummulative_return
```

```python
dropedna = cummulative_return.dropna()
annualized_return = (1 + cummulative_return.iloc[-1])**(252/len(dropedna))-1
annualized_return
```

```python
daily_volatility = daily_return.std()
annualized_volatility = daily_volatility * math.sqrt(252)
annualized_volatility
```

```python
sharpe = (annualized_return - 0.05)/annualized_volatility
sharpe
```

```python
print(f'Ação com maior retorno anualizado \n{annualized_return.sort_values(ascending=False)}')
print('-----')
print(f'Ação com maior indice Sharpe \n{sharpe.sort_values(ascending=False)}')
```
